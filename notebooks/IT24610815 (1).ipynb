```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')
file_path = '/content/drive/MyDrive/stores_sales_forecasting.csv'

# Load the dataset with the correct encoding
df = pd.read_csv(file_path, encoding='latin1')

print("--- Member 1: Handling Missing Data ---")
print("Original dataset shape:", df.shape)

# Identify and remove rows with missing values
df_cleaned = df.dropna()

print("Dataset shape after cleaning:", df_cleaned.shape)

# Visualization: Compare dataset size
original_rows = len(df)
cleaned_rows = len(df_cleaned)
plt.figure(figsize=(8, 6))
plt.bar(['Original', 'Cleaned'], [original_rows, cleaned_rows], color=['skyblue', 'salmon'])
plt.title('Comparison of Dataset Size Before and After Handling Missing Data')
plt.xlabel('Dataset State')
plt.ylabel('Number of Rows')
plt.show()

# Export the cleaned data for the next member
df_cleaned.to_csv('/content/drive/MyDrive/stores_sales_forecasting_cleaned_1.csv', index=False)
```
